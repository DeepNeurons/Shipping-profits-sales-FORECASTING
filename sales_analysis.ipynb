```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data = pd.read_excel('sales.xlsx')
```

```python
df = pd.DataFrame(data)
```

```python
df["Order Date"] = pd.to_datetime(df["Order Date"])
```

```python
df = df.sort_values(by="Order Date")
```

```python
df.head(3)
```

```python
print(len(df['Order ID']),len(df['Order ID'].unique()))
```

```python
print(len(df['Ship Mode']),len(df['Ship Mode'].unique()))
```

## Shipping comapnies

```python


Shipping_Modes = ['Delivery Truck','Regular Air','Express Air']

    
```

## Sales by shipping comapnie

```python
##
## Plot Shipping companies Sales
Sales = []
for sh in Shipping_Modes:
    Sales.append(df.loc[df['Ship Mode'] == sh,'Sales'].sum())
fig = plt.figure(figsize=(10,5))
plt.bar(Shipping_Modes,Sales,color='blue',
       width = 0.4)
plt.xlabel('Shipping companies')
plt.ylabel('Sales')
plt.title('Sales per Shipping companie')
plt.show()
```

## Profit by shipping companie

```python
##
##Plot Shippinf companies profits
profits = []
for sh in Shipping_Modes:
        profits.append(df.loc[df['Ship Mode'] == sh, 'Profit'].sum())
fig = plt.figure(figsize = (10, 5))
 
# creating the bar plot
plt.bar(Shipping_Modes, profits, color ='blue',
        width = 0.4)
 
plt.xlabel("Shipping companies")
plt.ylabel("Profits")
plt.title("Shipping companies profit")
plt.show()
```

```python
## Shipping over time for each companie

    
df_ = df.set_index('Order Date')
sns.set(rc={'figure.figsize':(15, 6)})
df_['Sales'].where(df_['Ship Mode'] == 'Delivery Truck').plot(linewidth=0.4,color='red')
plt.title('Delivery Truck sales over time')
plt.show()
    
```
